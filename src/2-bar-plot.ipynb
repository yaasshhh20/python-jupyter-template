```python
# TODO: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
to = 10  # User seems to have missed defining the starting x value; assuming start at 0.
x = np.arange(0, to+1)
y = x**2

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customization for Line Plot
plt.title('Line Plot of y = x^2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.legend()
plt.show()

# Bar Plot Data
categories = ['A', 'B', 'C']
values = [7, 5, 3]
colors = ['red', 'green', 'blue']

# Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)

# Customization for Bar Plot
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
