### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

try:
    # Ensure that x and y have the same length
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")

    # Create the scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', s=100, label="Data Points")

    # Annotate each point with its (x, y) value
    for i, j in zip(x, y):
        plt.text(i, j, f'({i}, {j})', fontsize=12, ha='right', va='bottom')

    # Add title and labels
    plt.title("Scatter Plot with Annotations", fontsize=16)
    plt.xlabel("X-axis", fontsize=14)
    plt.ylabel("Y-axis", fontsize=14)

    # Display the plot
    plt.legend()
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
