### Exercise 6: Histogram Customization

```python
# TODO: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

try:
    # Check if the data is a list and contains numeric values
    if not isinstance(data, list):
        raise TypeError("Data must be a list.")
    
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("All elements in the data list must be numeric (int or float).")
    
    # Create the histogram
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=5, color='skyblue', edgecolor='black')

    # Add title and labels
    plt.title("Histogram of Data", fontsize=16)
    plt.xlabel("Values", fontsize=14)
    plt.ylabel("Frequency", fontsize=14)

    # Display the plot
    plt.show()

except TypeError as e:
    print(f"Error: {e}")
except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
