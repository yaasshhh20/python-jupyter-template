### Exercise 4: Pie Chart with Percentages

```python
# TODO: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'skyblue', 'lightgreen', 'coral']

try:
    # Ensure that labels and sizes have the same length
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must have the same length.")
    
    # Ensure that sizes are non-negative
    if any(size < 0 for size in sizes):
        raise ValueError("Sizes must be non-negative values.")

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',  # Format to display percentages
        startangle=140,     # Rotate the start angle
        wedgeprops={'edgecolor': 'black'}  # Add a border to the wedges
    )

    # Add a title
    plt.title("Programming Language Popularity", fontsize=16)

    # Display the chart
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected erroroccurred:{e}")
```
